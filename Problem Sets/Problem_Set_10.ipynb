Using the following formula, we estimate $f''(x)$:

$$f''(x) \approx \frac{f(x+h) - 2f(x) + f(x - h)}{h^2}$$

```python
import numpy as np

x = np.pi / 6
f = lambda x: np.sin(x)

error = lambda x: np.abs(-0.5 - x)

def approximate(f, x, h):
  a = (f(x + h) - 2 * f(x) + f(x - h)) / h**2
  return (a, error(a))

print("True value of f\"(pi/6): -0.5")

print("\n(Approximate value of f\"(x), error) when")
print("h = 0.2:", approximate(f, x, 0.2))
print("h = 0.1:", approximate(f, x, 0.1))
print("h = 0.05:", approximate(f, x, 0.05))
```

As shown in the derivation above, after one step of Richardson extrapolation, the error should be of order $h^6$; after two steps, it should be of the order $h^8$. 

We now implement this result from the first step of Richardson extrapolation below. 

```python
def one_step(f, x, h):
  def phi(h):
    return (f(x + h) - 2 * f(x) + f(x - h))/h**2
  a = (4/3) * phi(h / 2) - (1/3) * phi(h)
  return (a, error(a))

print("(Approximate value of f\"(x), error) with Richarson extrapolation when")
print("h = 0.2:", one_step(f, x, 0.2))
print("h = 0.1:", one_step(f, x, 0.1))
```

Finally, we implement the result from the second step of Richardson extrapolation below. 

```python
def two_step(f, x, h):
  def phi(h):
    return one_step(f, x, h)[0]
  
  a = (64/63) * phi(h / 2) - (1/63) * phi(h)
  return (a, error(a))

print("(Approximate value of f\"(x), error) with Richarson extrapolation when")
print("h = 0.2:", two_step(f, x, 0.2))
print("h = 0.1:", two_step(f, x, 0.1))
```
