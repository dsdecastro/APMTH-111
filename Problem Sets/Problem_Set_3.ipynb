## Problem 24

```python
import numpy as np

# Define gaussElim
def gaussElim(A, b):
  n = len(b)

  # Elimination
  for k in range(0, n-1):
    for i in range(k+1, n):
      if A[i,k] != 0.0:
        lam = A[i,k] / A[k,k]
        A[i,k+1:n] -= lam*A[k,k+1:n]
        b[i] -= lam*b[k]

  # Back substitution
  for k in range(n-1,-1,-1):
    b[k] = (b[k] - np.dot(A[k,k+1:n],b[k+1:n]))/A[k,k]
  
  return b

# Define A and b
A = np.array([[5+1j, 5+2j, -5+3j, 6-3j],
              [5+2j, 7-2j, 8-1j, -1+3j],
              [-5+3j, 8-1j, -3-3j, 2+2j],
              [6-3j, -1+3j, 2+2j, 8+14j]])

b = np.array([15-35j, 2+10j, -2-34j, 8+14j])

print(gaussElim(A, b))
```

```python
import numpy as np

# Define gaussElim

def gaussElim(A, B):
  n = len(A)

  # Elimination
  for k in range(0, n-1):
    for i in range(k+1, n):
      if A[i,k] != 0.0:
        lam = A[i,k] / A[k,k]
        A[i,k+1:n] -= lam*A[k,k+1:n]
        B[i,:] -= lam*B[k,:]

  # Back substitution
  for k in range(n-1,-1,-1):
    for c in range(0, len(B[0])):
      B[k,c] = (B[k,c] - np.dot(A[k,k+1:n], B[k+1:n,c]))/A[k,k]

  return B

# Define A and B
A = np.array([[2., -1., 0.],
              [-1., 2., -1.],
              [0., -1., 2.]])
B = np.array([[1., 0., 0.],
              [0., 1., 0],
              [0., 0., 1.]])

print(gaussElim(A, B))
```
