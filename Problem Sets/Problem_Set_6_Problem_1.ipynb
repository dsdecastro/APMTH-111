## 2) 

Based on the divided differences table above, the Newton interpolation polynomial is 

$$p(x) = -53 + 26(x+4) - 7(x+4)(x+2) + (x+4)(x+2)(x+1)$$

This polynomial is of degree 3. 

# 3)

```python
import numpy as np
from scipy.interpolate import lagrange

# Find and print the Lagrange form of the polynomial
poly_L = lagrange(x_val, y_val)
print("Lagrange form: q(x) = ")
print(poly_L, "\n")
```

By executing the code above, we have found the Lagrange form of the interpolation polynomial to be 

$$q(x) = -1.388 \times 10^{-17}x^5 + 5.551\times 10^{-17}x^4 + x^3 - 4.441 \times 10^{-16}x^2 - 2x + 3$$

The graph produced by the code below demonstrates that this polynomial is equivalent to the Newton form found in part (2). 


```python
import matplotlib.pyplot as plt

# Configure plot
fig, ax = plt.subplots()

# Create range, array of x values, and array of f(x) values
x = np.arange(-5, 5, 0.1)
x_val = np.array([-4,-2,-1,1,3,4])
y_val = np.array([-53,-1,4,2,24,59])

# Define both polynomials
poly_n = -53 + 26 * (x+4) - 7 * (x+4) * (x+2) + (x+4) * (x+2) * (x+1) 

# Plot the points and the polynomials
ax.plot(x_val, y_val, 'bo')
ax.plot(x, poly_n, color="turquoise", marker=None, linestyle="solid", \
        label="p(x)")
ax.plot(x, poly_L(x), color="navy", marker=None, linestyle="dashed", \
        label="q(x)")

# Add legend and axis labels, and draw graph
ax.legend()
plt.draw()
```

```python
# Configure plot
fig, ax = plt.subplots()

# Create range, array of x values, and array of f(x) values
x = np.arange(-5, 5, 0.1)
x_val = np.array([-4,-2,-1,1,3,4])
y_val = np.array([-53,-1,4,2,24,59])

# Define the interpolation polynomial
poly = -53 + 26 * (x+4) - 7 * (x+4) * (x+2) + (x+4) * (x+2) * (x+1) 

# Plot the points and the polynomial
ax.plot(x_val, y_val, 'bo')
ax.plot(x, poly, color="navy", marker=None, linestyle="solid", \
        label="p(x)")

# Add legend, and draw graph
ax.legend()
plt.draw()
```
